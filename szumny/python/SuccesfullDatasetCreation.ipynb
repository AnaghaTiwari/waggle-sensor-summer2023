```python
# xml annotation to csv
import cv2
import os
import pandas as pd
import xml.etree.ElementTree as ET



def convert_xml2csv(xml_file):
    columns = ["filename", "folder", "type", "occluded", "date", "xmin", "ymin", "xmax", "ymax"]
    xml_data = []
    xml_df=pd.DataFrame()
    try:
        file_parse = ET.parse(xml_file)
       
        root = file_parse.getroot()
        
        for i in root.iter('object'):

            # xmin = 10000
            # xmax = 0
            # ymin = 10000
            # ymax = 0

            # for y in i[8]:
            #     for b in y:
                    
            #         if int(y[0].text) > ymax:
            #             ymax = int(y[0].text)
                    
            #         if int(y[0].text) < ymin:
            #             ymin = int(y[0].text)
                    
            #         if int(y[1].text) > xmax:
            #             xmax = int(y[1].text)
                    
            #         if int(y[1].text) < xmin:
            #             xmin = int(y[1].text)
            

            if((i[0].text != "smoke plume")&(i[0].text != "no smoke")&(i[0].text != "abc")&(i[0].text != "bp") & (i[0].text != "sp")&(i[0].text != "small plume")
               | (i[0].text == "smoke")):
                xmin = 10000
                xmax = 0
                ymin = 10000
                ymax = 0

                for y in i[9]:
                    for b in y:
                        if int(y[1].text) > ymax:
                            ymax = int(y[1].text)
                        
                        if int(y[1].text) < ymin:
                            ymin = int(y[1].text)
                        
                        if int(y[0].text) > xmax:
                            xmax = int(y[0].text)
                        
                        if int(y[0].text) < xmin:
                            xmin = int(y[0].text)
                value = (root.find('filename').text,
                        root.find('folder').text,
                        i[0].text, #smoke
                        i[3].text,#occluded
                        i[6].text, #date
                        xmin,
                        ymin,
                        xmax,
                        ymax
                )
            
                xml_data.append(value)

            xml_df = pd.DataFrame(xml_data, columns=columns)
    except Exception as e:
        print('xml conversion failed:{}'.format(e))
        return pd.DataFrame(columns=columns)
    return xml_df
```

```python
import os

fulldata = pd.DataFrame()

directory = '/home/jszumny/Documents/copy of smoke data/hpwren/data/'
for filename in os.listdir(directory):

    if "._" not in filename and filename.__contains__("smoke_detection"):

        for file in os.listdir(directory + str(filename) +'/'):

            if "._" not in file and ".D" not in file:

                for xml in os.listdir(directory + str(filename) + "/" + str(file)):

                    if "._" not in xml:
                        filepath = os.path.join(directory, filename, file, xml)
                        fulldata = pd.concat([fulldata, convert_xml2csv(filepath)], ignore_index=True)
                        
                            
                            
```
