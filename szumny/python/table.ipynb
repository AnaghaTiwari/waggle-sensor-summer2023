```python
import pandas as pd
```

```python
smoke = pd.read_csv("/home/jszumny/Documents/copy of smoke data/hpwren/python/smoke-invert-resnet50.csv")
nosmoke = pd.read_csv("/home/jszumny/Documents/copy of smoke data/hpwren/python/nosmoke-invert-resnet50.csv")

# smoke = pd.read_csv("/home/jszumny/Downloads/smoke-UCSD.csv")
# nosmoke = pd.read_csv("/home/jszumny/Downloads/no-smoke-UCSD.csv")

smoke["smoke"] = True
smoke

nosmoke["smoke"] = False

full = pd.concat([smoke,nosmoke])
full
```

```python
#TN
tn = len(full[((full['class'] != 'smoke') & (full.smoke == False)) | ((full['class'] != 'smoke') & (full.smoke == True) & ((full.xtile != 1) | (full.xtile != 2)) & (full.ytile !=0))])

#FP
fp = len(full[((full['class'] == 'smoke') & (full.smoke == False)) | ((full['class'] == 'smoke') & (full.smoke == True) & ((full.xtile != 1) | (full.xtile != 2)) & (full.ytile !=0))])

#TP
tp = len(full[(full['smoke'] == True) & (full.ytile == 0) & ((full.xtile == 1) | (full.xtile == 2)) & (full['class'] == "smoke")])

#FN
fn = len(full[(full['smoke'] == True) & (full.ytile == 0) & ((full.xtile == 1) | (full.xtile == 2)) & (full['class'] != "smoke")])

# Rates
tpr = tp / (tp+fn)
print( "True Positive Rate: ",tpr)

fnr = fn / (tp + fn)
print( "False Negative Rate: ",fnr)

tnr = tn / (tn +fp)
print( "True Negative Rate: ",tnr)

fpr = fp / (fp + tn)
print( "False Positive Rate: ",fpr)
```
