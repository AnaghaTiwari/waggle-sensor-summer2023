```python
import torch
from ultralytics import YOLO
import numpy as np
```

```python
model = YOLO(model = 'best_s2.pt')

results = model.predict('./../../rideshare/new_rideshare_super4/', save = True)
length = 0
sum = 0
confs_super = []
confs_super_list = []
for result in results:
    length += len(result.boxes.conf)
    sum +=torch.sum(result.boxes.conf).item()
    confs_super = confs_super + result.boxes.conf.tolist()
    confs_super_list.append(result.boxes.conf.tolist())
print(sum/length)
print(confs_super)
```

```python
model = YOLO(model = 'best_s2.pt')

results = model.predict('./../../rideshare/new_rideshare/', save = True)
length = 0
sum = 0
confs = []
confs_list = []
for result in results:
    print(result)
    length += len(result.boxes.conf)
    sum +=torch.sum(result.boxes.conf).item()
    confs = confs + result.boxes.conf.tolist()
    confs_list.append(result.boxes.conf.tolist())
print(sum/length)
print(confs)
```

```python
import matplotlib.pyplot as plt


# Create a figure and axis
fig, ax = plt.subplots()
binwidth = 0.2
bin_edges = np.arange(0.7, 0.95, 0.01) 
# Plot histograms for both lists
ax.hist(confs_super, bins=bin_edges, alpha=0.5, label='Super resolution')
ax.hist(confs, bins=bin_edges, alpha=0.5, label='Normal Resolution')

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Confidence of predictions')
ax.legend()

# Show the plot
plt.show()
```
